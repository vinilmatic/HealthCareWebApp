## Solving it using Pandas Library.

```python
import pandas as pd
```

```python
df = pd.read_csv('HealthData(in).csv')
df
```

```python
#(Q) Calculate the average age and BMI of the individuals in the dataset.
avg_age = df['age'].mean()
avg_BMI = df['bmi'].mean()

print('Average age: ', avg_age)
print('Average BMI: ', avg_BMI)
```

```python
#(Q)Determine how many smokers and non-smokers are in the dataset, and what percentage of the total each group represents
breakdown = df['smoker'].value_counts()
print(breakdown)

smokers = round((breakdown['yes']/len(df))*100, 2)
non_smokers = round((breakdown['no']/len(df))*100, 2)

print('% Smokers: ', smokers)
print('% Non-smokers: ', non_smokers)
```

```python
#(Q)Find the minimum, maximum, and median insurance charges in the dataset.
minimum = df['charges'].min()
maximum = df['charges'].max()
median = df['charges'].median()

print('Minimum insurance charges: ', minimum)
print('Maximum insurance charges: ', maximum)
print('Median insurance charges: ', median)
```

```python
#(Q) Create a new column called 'age_group' that categorizes individuals into 'Young' (18-35), 'Middle-aged' (36-55), and 'Senior' (56+), then calculate the average charges for each age group.
def categorize_age(age):
    if age <= 35:
        return('Young')
    elif age >= 56:
        return('Old')
    else:
        return('Middle-aged')

df['age_group'] = df['age'].apply(categorize_age)
df
```
