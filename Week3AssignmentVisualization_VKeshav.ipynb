```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HealthData(in).csv')
df
```

```python
# Plot a scatter plot between BMI and insurance charges. 
#Write down your observations about the relationship between these two variables.
#df.plot(kind='scatter', x='bmi', y='charges')
plt.figure(figsize=(10,6))
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df, alpha=0.7)
plt.title('BMI vs Insurance Charges', fontsize=14)
plt.xlabel('BMI')
plt.ylabel('Insurance Charges ($)')
plt.show()
```

Observation: Smokers have higher insurance charges than non smokers regardless of their BMI. For example, a smoker with a low BMI has a higher insurance charge than a non smoker with a high BMI. There does not appear to be a distinct relationship between BMI and insurance charges.

```python
sns.displot(df, x='charges')
plt.title('Histogram showing insurance charges', fontsize=14)
plt.show()
```

Observation: The graph is skewed right. Majority of people pay less than 15 000.

```python
sns.displot(df, x='smoker')
plt.title('Histogram of number of smokers', fontsize=14)
plt.show()
```

Observation: There are significantly more non-smokers than smokers.

```python
sns.displot(df, x='age')
plt.title('Historgram of ages', fontsize=14)
plt.show()
```

Observation: The graph somewhat resembles a normal distribution. It is evenly split across all ages with the exception of ages 18-22.
