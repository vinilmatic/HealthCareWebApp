```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#Loading CSV File into our Dataframe
df =  pd.read_csv('HealthData.csv')
```

## <font color='green'> 2D Scatter Plot

```python
df.plot(kind='scatter', x='age', y='charges') ;
plt.show() ;
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='charges', hue='smoker', data=df, palette='viridis', alpha=0.7)
plt.title('Age vs Charges with Smoking Status', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Charges ($)')
plt.show()
```

## <font color='green'> Pair-Plot 

```python
sns.pairplot(df, hue= 'smoker', diag_kind='kde')
```

### Observation : The scatter plot  between  "age" and  "charges"  showing that.  Insurance charges generally increase with age.
### Smoking seems to have high impact on charges.


## <font color='green'> Histograms

```python
sns.displot(df, x="bmi" );
```

## Observation The distributions of the BMI is almost normal with a slight right skew.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='region', data=df, palette='Set2')
plt.title('Distribution of Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

## Region is Evenly Distributed 

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df, palette='Set2')
plt.title('Distribution of Sex', fontsize=14)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```
